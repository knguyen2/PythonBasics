```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
reviews = pd.read_csv('winemag-data_first150k.csv', index_col=0)
```

```python
reviews.sample(5)
```

```python
#Filter for all prices that are under $100
reviews[reviews['price'] < 100].sample(100).plot(kind='scatter', x='price', y='points')
#Scatter plot is used to see/compare the relationship between 2 variables
```

```python
#Use Hexbin instead of scatter
reviews[reviews['price'] < 100].plot(kind='hexbin', x='price', y='points')
```

```python
reviews[reviews['price'] < 100].plot(kind='hexbin', x='price', y='points', gridsize=15)
```

```python
#Filter price < $100 and price > $20

reviews[(reviews.price < 100) & (reviews.points > 20)].plot(kind='box')

#Outliers are the circles outside the box 
```

```python
#Creating a subplot example with 2 rows x 1 column

fig, axarr = plt.subplots(2,1, figsize=(10,8))

reviews.points.value_counts().sort_index().plot(kind='bar', ax=axarr[0])

reviews.province.value_counts().head(20).plot(kind='bar', ax=axarr[1])
```
