Pandas DataFrame from numpy.array

```python
import numpy as np
import pandas as pd
```

```python
data = np.array([['', 'Col1', 'Col2'],
                ['Row1',1,2],
                ['Row2',3,4]])
```

```python
df = pd.DataFrame(data[1:,1:],
index = data[1:,0],
columns = data[0,1:])
```

```python
# Declaring variable col_names with data and pass it into the colums
col_names = ['Alabama','Bahamas','Cali','Denmark']
my_second_df = pd.DataFrame(data = [[1,2,3,5],[4,5,5,6],[4,2,5,6]],
                     index=range(0,3), columns=col_names)
```

```python
my_second_df
```

positional reference (.iloc[]) or getting data by using names (.loc[])

```python
# return row 1 from df

my_second_df.iloc[[1]]
```

```python
#get column Bahamas using position ref

my_second_df.iloc[:,[1]]
```

```python
#get data from both column 1 and column 2 using positional ref
# my_second_df.iloc[:,[0,1]]
my_second_df.iloc[:,0:2]
```

```python
# get column Cali using names
my_second_df.loc[:,['Cali']]
```

```python
#subset Bahamas and Cali using names

my_second_df.loc[:,['Bahamas', 'Cali']]
```

```python
#get top 2 tows for all columns - names
# my_second_df.loc[:1,['Alabama','Bahamas', 'Cali', 'Denmark']]
my_second_df.loc[:1,:]
```

```python
#get the top 2 rows for all columns - using positional ref
my_second_df.iloc[:2,0:4]
```

```python
#get the last 2 rows of Cali and Denmark - positional ref

my_second_df.iloc[-2:,[2,3]]
```

```python
#get the last 2 rows of Cali and Denmark - names

my_second_df.loc[[1,2], ['Cali', 'Denmark']]
```

```python
my_second_df
```

```python
#change row[2] (3rd row) data - positional ref

my_second_df.iloc[[2]]
```
